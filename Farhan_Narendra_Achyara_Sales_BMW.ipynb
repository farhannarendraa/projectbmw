## 1.1 Latar Belakang

BMW merupakan perusahaan multinasional berbasis di Jerman yang memiliki kelas mewah dan dibeli oleh konsumen kaya dan menengah ke atas. ## 1.2 Definisikan Problem Statement menggunakan SMART

- Specific: Meningkatkan penjualan mobil BMW dari berbagai wilayah. Bersaing dengan kompetitor terutama dari Tiongkok

- Achievable: Strategi periklanan harus lebih baik. Produksi mobil di wilayah yang biaya produksi lebih kecil, bikin produk yang lebih sesuai dengan keinginan konsumen
- Relevant: Keuangan BMW menjadi lebih stabil setelah beberapa tahun tertekan pasca pandemi COVID-19
- Time-Bound: Waktu yang diperlukan agar penjualan dapat meningkat kembali dalam 1 - 2 tahun.

```python
df_bmw_time['Revenue'].mean()
df_bmw_time['Revenue'].median()
df_bmw_time['Revenue'].std()
df_bmw_time['Revenue'].var()

print("Rata-rata pendapatan mobil terjual:", df_bmw_time['Revenue'].mean(), "USD")
print("Median pendapatan mobil terjual:", df_bmw_time['Revenue'].median(), "USD")
print("Simpangan baku pendapatan mobil terjual:", df_bmw_time['Revenue'].std(), "USD")
print("Ragam pendapatan mobil terjual:", df_bmw_time['Revenue'].var())
```

```python
df_bmw_time['Revenue'].agg(['skew', 'kurtosis'])
```

Dari data pendapatan dari penjualan mobil, sudah sangat jelas terlihat bahwa dari rata-rata dan median pendapatan dari penjualan mobil mirip-mirip sehingga data pendapatan dari penjualan mobil dalam USD dipastikan normal. Diperkuat dengan perhitungan skewness dan kurtosis yang mendekati 0. 

Langkah selanjutnya adalah menentukan selang kepercayaan dari rata-rata pendapatan dari mobil BMW. Karena dari ukuran tendensi sentral sudah menunjukkan bahwa sebaran sudah normal, dapat dilakukan perhitungan selang kepercayaan dari sebaran normal juga. 

```python
std = df_bmw_time['Revenue'].std()
N = len(df_bmw_time['Revenue'])
batas_bawah, batas_atas = stats.norm.interval(0.95,loc=df_bmw_time['Revenue'].mean(),scale=std/np.sqrt(N))
print('Batas bawah:', batas_bawah, "USD")
print('Batas atas:', batas_atas, "USD")
```

Dari perhitungan selang kepercayaan, dapat disimpulkan bahwa potensi pendapatan perusahaan BMW berkisar antara USD 378.052.809 sampai USD 384.798.770 per order dengan selang kepercayaan 95%. Hal tersebut menunjukkan bahwa potensi pendapatan maksimum dari pendapatan per order sebesar USD 384.798.770. 

Untuk menentukan apakah rata-rata harga mobil di Asia berbeda dengan harga mobil di wilayah lainnya. Perlu untuk dilakukan uji hipotesis dua sampel dan dua arah dengan:

- **H0**: Tidak ada perbedaan antara rata-rata harga mobil BMW di Asia dengan wilayah lainnya
- **H1**: Terdapat perbedaan antara rata-rata harga mobil BMW di Asia dengan wilayah lainnya

```python
# Ambil kategori Asia dalam kolom region

df_bmw_time_asia = df_bmw_time[df_bmw_time['Region'].isin(['Asia'])]
df_bmw_time_asia
```

```python
# Ambil region selain Asia dari kolom region

df_bmw_time_notasia = df_bmw_time[-df_bmw_time['Region'].isin(['Asia'])]
```

```python
# Uji hipotesis perbandingan harga mobil di Asia dan di luar Asia

t_stat, p_val = stats.ttest_ind(df_bmw_time_asia['Price_USD'], df_bmw_time_notasia['Price_USD'])
print('T-Statistic:',t_stat)
print('P-value:',p_val) 
```

Dari hasil uji hipotesis, dapat disimpulkan bahwa hasil uji menolak **H0** dengan titik kritis p-value 5%. Terdaoat perbedaan antara rata-rata harga mobil di Asia dengan wilayah lainnya dengan taraf nyata 5%. Ada faktor-faktor yang menjadi penyebab harga berbeda seperti bahan baku, biaya pengiriman, biaya karyawan, dll.
